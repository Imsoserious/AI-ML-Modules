# 선형회귀/경사하강법 실습

```python
import numpy as np
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential

x_train = np.array([1, 2, 3]) # 인풋 데이터
y_train = np.array([1, 2, 3]) # 결과값(정답)

model = Sequential() # 모델 객체 생성, layer를 추가 가능
model.add(Dense(1, input_dim=1)) # Dense는 퍼셉트론, 층 하나, 인풋 데이터의 차원은 1차원

sgd = SGD(lr=0.10) # 경사하강법(학습률=0.1)

# compile 메소드로 함수 설정
model.compile(loss='mse', optimizer=sgd) # 비용함수는 mse(mean square error), optimizer는 경사하강법

# prints summary of the model to the terminal
model.summary()

model.fit(x_train, y_train, epochs=200)

y_predict = model.predict(np.array([5])) # 예상값
print(y_predict)
```

```python
import numpy as np
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense # Activation은 활성화 함수
from tensorflow.keras.optimizers import RMSprop

# 데이터 입력
x_train = np.array([2,4,5,7,8,10])
y_train = np.array([0,0,0,1,1,1])

# 모델 생성
model = Sequential()

# 레이어 추가 및 인풋데이터의 속성 수 입력
model.add(Dense(1, input_dim=1))

# 모델에 활성화함수 추가
model.add(Activation("sigmoid"))

# 학습율 설정
#rmsprop = RMSprop(lr=1e-2)
sgd = SGD(lr=0.01)

# 비용함수와 학습률 입력
#model.compile(loss='binary_crossentropy', optimizer=rmsprop)
model.compile(loss='binary_crossentropy', optimizer=sgd)

# prints summary of the model to the terminal
model.summary()

# 모델 학습
model.fit(x_train, y_train, epochs=500)

# 테스트
y_predict = model.predict(np.array([3]))
print(y_predict)

# 결과값을 시그모이드 적용해서 0,1로 반환
model.predict_classes(np.array([3]))
```

# 1. 데이터 준비

```python
from tensorflow.keras.datasets import boston_housing

# traindata= 속성, train_targets=클래스(정답)
(train_data, train_targets), (test_data, test_targets) = boston_housing.load_data() # 입력순서 꼭 지키기
```

```python
print(train_data.shape)
print(test_data.shape)
```

# 2. 데이터 전처리

```python
# 각 속성 표준화
mean = train_data.mean(axis=0) # 속성별로 평균 구하기
train_data-=mean
std = train_data.std(axis=0) # 속성별로 표준편자 구하기
train_data/=std

# 테스트 셋도 훈련 셋에서 계산된 mean과 std를 동일하게 사용함
# 만약 각각 정규화를 수행해버리면, 학습 결과 적용이 불가능해짐
# (테스트 데이터는 원칙적으로 모르는 데이터이기 때문에 전처리도 훈련셋과 똑같이 해야 함.)

test_data -=mean
test_data /=std
```

# 3. 네트워크 정의 

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

def build_model():
    model = Sequential()
    # 층 설정: 1층(64 unit)--2층(64 unit)--3층(1 unit)
    model.add(Dense(64, activation='relu', input_shape=(train_data.shape[1],)))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(1)) # 회귀 모델 (출력층에 활성화 함수 없음)
    
    model.compile(optimizer='rmsprop', loss='mse', metrics=['mae']) # rmspop도 경사하강법인데 sgd보다 성능 좋음
    # metrics=['mae']는 mean absolute error 실제 집가격과 예측된 집가격의 절대값을 비교하여 평가
    # mse는 나쁜 예측에 패널치를 주기 때문에 학습에 좋고, mae는 차이를 그대로 비교하므로 평가에 좋다.
    
    return model
```

# 4. (K-겹 교차 검증을 사용한) 모델 학습
#### - 모델이 단순하면서도 데이터 크기가 작을 때 사용

```python
import numpy as np

### 목적: 평균 성능 측정 ###

k = 4 # 0~3번 폴드
num_val_samples = len(train_data) // k
# 에폭이란?
# 데이터 1회전1
num_epochs = 100
all_scores = [] # 각 폴드의 스코어 저장하는 빈 리스트 생성

for i in range(k):
    print('>> fold ', i)
    
    # 훈련셋과 검증셋으로 나누기
    val_data = train_data[i * num_val_samples: (i+1) * num_val_samples]
    val_targets = train_targets[i * num_val_samples: (i+1) * num_val_samples]
    
    # 떨어져 있는 훈련셋 조각을 하나로 붙임 ex) t=[t1,t2,v1,t3] -> concatenate -> t=[t1,t2,t3],v=v1
    # concatenate: 이어 붙이기
    partial_train_data = np.concatenate([train_data[:i*num_val_samples],
                                         train_data[(i+1) * num_val_samples:]],
                                        axis=0)
    partial_train_targets = np.concatenate([train_targets[:i*num_val_samples],
                                            train_targets[(i+1) * num_val_samples:]],
                                           axis=0)
    
    model = build_model() # 위의 네트워크 정의 함수
    
    # 모델 학습
    
    # verbose=0이면 세부결과 안 나옴, 1이면 나옴
    
    # batch size=1 이므로 각 인스턴스에 대해 경사하강을 할 것임
    # batch 란: 작업량을 나누어서 실행하는 것. 
    # 데이터가 1TB인데 RAM은 8GB이므로 한 번에 올려서 작업이 불가능함. 그래서 나누어서 연속적으로 작업을 진행
    # batch size는 컴퓨터 구조상 2의 배수로 정해주는 게 제일 효율이 좋음
    
    model.fit(partial_train_data, 
              partial_train_targets,
              epochs = num_epochs,
              batch_size = 1,
              verbose = 0)
    
    # 모델 평가
    val_mse, val_mae = model.evaluate(val_data, val_targets, verbose=0) # return: loss 결과, metrics=mae 결과가 리스트 형태로 나옴
    all_scores.append(val_mae)

print('finished!!')
```

```python
print(all_scores) # 4개 폴드에서 나온 mae
mean = np.mean(all_scores)
stddev = np.std(all_scores)

# 폴드의 평균, 표준편차 
# 정확도와 치우침 등 허용오차 한계선은 도메인과 과제 목표에 따라 달라짐
print(u'%.2f \u00B1(%.2f)'% (mean,stddev))
```

```python
# 에폭을 500으로 늘리고 평가

### 목적: 과적합되는 학습횟수 찾기 ###

num_epochs = 500
all_mae_histories = []

for i in range(k):
    print('>> fold ', i)
    val_data = train_data[i * num_val_samples: (i+1) * num_val_samples]
    val_targets = train_targets[i * num_val_samples: (i+1) * num_val_samples]
    
    partial_train_data = np.concatenate([train_data[:i*num_val_samples],
                                         train_data[(i+1) * num_val_samples:]],
                                        axis=0)
    partial_train_targets = np.concatenate([train_targets[:i*num_val_samples],
                                            train_targets[(i+1) * num_val_samples:]],
                                           axis=0)
    
    model = build_model()
    # fit 메소드는 모델 학습 후의 histroy를 보관함.
    hist = model.fit(partial_train_data, 
                     partial_train_targets,
                     epochs = num_epochs,
                     batch_size = 1,
                     verbose = 0,
                     validation_data = (val_data, val_targets) # validation_data=()라는 변수 추기
    
    mae_history = hist.history['val_mean_absolute_error'] # validation_Data 변수 추가했기에 저장된 히스토리임
    all_mae_histories.append(mae_history)
    
print('finished!!')
```

```python
# history의 mae는 한 에폭 당 나온 mae. mean_mae_history는 4개 폴드의 각 i번째 에폭의 평균 즉, 500개
```

```python
'''
시계열 신호에 지수 이동 평균(exponential moving averge) 필터링
'''
def smooth_curve(points, factor=0.9):
    smoothed_points = []
    for point in points:
        if smoothed_points:
            prev = smoothed_points[-1]
            smoothed_points.append(prev*factor + point*(1-factor))
        else:
            smoothed_points.append(point)
    
    return smoothed_points

smooth_mae_history = smooth_curve(average_mae_history[10:])

plt.plot(range(1, len(smooth_mae_history) + 1), smooth_mae_history)
plt.xlabel('Epochs')
plt.ylabel('Validation MAE')
plt.show()
```
