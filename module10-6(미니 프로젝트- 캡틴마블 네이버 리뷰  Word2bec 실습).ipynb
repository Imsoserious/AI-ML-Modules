```python
def get_data(movie_url, total_count):

    text_list = list()
    
    for i in range(1, int(total_count / 10) + 1):
            url = movie_url + '&page=' + str(i)
            resp = requests.get(url)
            html = BeautifulSoup(resp.content, 'html.parser')
            score_result = html.find('div', {'class': 'score_result'})
            lis = score_result.findAll('li')
            
            for li in lis:  
                review_text = li.find('p').getText()
                # 네이버 영화 댓글에는 '관람객' 표시가 뜨는 경우가 있다. p  사이에 위치하여서 같이 크롤링 되었다.
                # 의미 없는 단어이기 때문에 댓글 첫 단어가 '관람객'인 경우 '관람객'을 지웠다.
                if(review_text[:3] == "관람객"):
                    review_text = review_text[3:]
                text_list.append(review_text)
            print("페이지 {}/{}".format(i,int(total_count / 10) + 1))
    return text_list
```

```python
from konlpy.tag import Hannanum
han = Hannanum()

result = []

for line in review:
    mallist = han.pos(line)
    r=[]
    for word in mallist:
        if not word[1] in ["J","E","S"]: 
            r.append(word[0])
    rl = (" ".join(r)).strip()
    result.append(rl)
```

```python
with open("Captain_Marvel.nip","w",encoding="utf-8") as fp:
    fp.write("\n".join(result))
```

```python
from gensim.models import word2vec

wData = word2vec.LineSentence("Captain_Marvel.nip")
wModel = word2vec.Word2Vec(wData, size=5000, window=10, min_count=2, sg=1)
wModel.save("Captain_Marvel.model")
print("Word2Vec Modeling finished")
```
