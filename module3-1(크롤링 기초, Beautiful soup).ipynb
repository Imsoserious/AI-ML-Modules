```python
from bs4 import BeautifulSoup
import urllib.request as req

url = 'https://finance.naver.com/marketindex/'
soup= BeautifulSoup(req.urlopen(url),'html.parser')

title = soup.select_one('#container > div.market_include > div > div.market1 > div.title > h2 > span')
print(title.string)
h1 = soup.select_one('#exchangeList > li.on > a.head.usd > h3').string
h2 = soup.select_one('#exchangeList > li.on > a.head.usd > div > span.value').string
print('{0} = {1}'.format(h1,h2))

h3 = soup.select('li > a> h3')
h4 = soup.select('li > a > div > span.value')
n= len(h3)
for i in range(n):
    print('{0} = {1}'.format(h3[i].string,h4[i].string))
```

```python
text = '''student-1-:jame
student-2-
student-3-
student-4-
lecturer-1-
'''
p = re.compile('^student')
result = p.findall(text)
print(result)

p = re.compile('^student.*')
result = p.findall(text)
print(result)

p = re.compile('^student.*', re.MULTILINE)
result = p.findall(text)
print(result)
```

```python
p = re.compile('[^c]') #  [^]는 ^ 뒤의 문자를 제외한 글자를 출력
result = p.findall('c22')
print(result)

p = re.compile('^c') #  ^는 뒤의 문자랑 같은 문자만 출력
result = p.findall('c22')
print(result)
```

```python
p = re.compile('[2.]') #  [.,+]는 따로따로 출력
result = p.findall('c22c22')
print(result)

p = re.compile('2.') #  [] 없으면 합쳐서 출력
result = p.findall('c22c22')
print(result)
```
