```python
x=np.arange(1,11)
y=x*5+np.random.randn(10)*5
y2=5+x*5+np.random.randn(10)*5

plt.plot(x,y,"p")
plt.plot(x,y2,"rd:")
plt.title("산점도 제목",fontdict={'fontsize': 25, "fontweight": "bold"},loc="left")
plt.xlabel("x축",labelpad=20)
plt.ylabel("y축",labelpad=15)
plt.legend(["남","여"],loc="lower right", fontsize='x-large')
plt.axis(xmin=0,xmax=15,ymin=0,ymax=100)
plt.minorticks_on()
plt.grid(b=True,which="major",axis="y")


plt.show()
```

```python
apt=pd.read_csv("C:/Users/Affinity/Desktop/수업/Python_Data/ch02_data/APTData.csv",header=0, skiprows=7,index_col=None,
                encoding="utf-8", na_values=[" ","."])
apt.Area
```

```python
area=pd.value_counts(apt["Area"])
plt.bar(area.index,area,bottom=10,color="red",edgecolor="b")
```

```python
plt.pie(area,labels=area.index,colors=["r","m","b","g","c","y"],explode=(0,0,0,0,0.5,0),
       autopct="%5.2f%%",shadow=True) # explode의 숫자는 원점과의 거리
plt.show()
```

```python
plt.boxplot([1,3,4,5,6,7,5,4,3,5,12],notch=True,vert=False,labels=["임대가"],whis=3)
plt.show()
```

```python
plt.boxplot(apt.Price,notch=True,vert=False,labels=["임대가"],whis=1,showmeans=True)
plt.show()
```

```python
# 히스토그램에서 급간 폭이 같으면 빈도 밀도 둘 다 가능, 다르면 밀도만 가능
plt.hist(apt.Price,bins=8,density=False,color="k")
```

```python
plt.plot(apt.Price,apt.FAR,"ro--")
plt.plot(apt.FAR,"bd--")
plt.title("산점도")
plt.legend(["임대가","FAR"])
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.subplot(221)

plt.hist(apt.Price,bins=8)
plt.subplot(224)
plt.hist(apt.FAR,bins=8)
plt.subplot(222,frameon=False)
plt.boxplot(apt.Price,labels=["임대가"],vert=False)
plt.subplot(223,frameon=False)
plt.boxplot(apt.FAR,vert=False,labels=["용적률"])
plt.show()
```

# plotly

```python
import plotly.graph_objs as go
import plotly as ply
```

```python
d= [go.Pie(text=area.index,values=area.values,opacity=0.7)]
layout=go.Layout(title="원그래프")
figure=go.Figure(data=d,layout=layout)
ply.offline.plot(figure,filename=\
                "C:/Users/Affinity/Desktop/수업/Python_Data/ch02_data/piechart.html")
```

```python
d2= [go.Bar(x=area.index,y=area.values)]
layout2=go.Layout(title="막대그래프")
figure=go.Figure(data=d2,layout=layout)
ply.offline.plot(figure,filename=\
                "C:/Users/Affinity/Desktop/수업/Python_Data/ch02_data/barchart.html")
```

```python
d3= [go.Box(x=area.index,y=area.values,boxmean=True,fillcolor="yellow",jitter=True,notched=True)]
layout2=go.Layout(title="그래프")
figure=go.Figure(data=d3,layout=layout)
ply.offline.plot(figure,filename=\
                "C:/Users/Affinity/Desktop/수업/Python_Data/ch02_data/boxchart.html")
```
