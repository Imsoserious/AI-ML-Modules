```python
va = re.compile('[a-z]+') # a to z까지 적어도 한 번 반복하는 문자패턴

# match: 글자 순으로 첫번째 패턴까지만 찾음
print(va.match('aaaaa'))
print(va.match('0a')) # 앞자ㅔva.match('0a') # match는 첫번째 글자부터 매치 여부를 판단,
print(va.match('a0b')) # 두번쨰 글자가 패턴에 없으므로 첫번째 문자 a만 매치로 판단
```

```python
va = re.compile('a.') # a to z까지 적어도 한 번 반복하는 문자패턴

# match: 글자 순으로 첫번째 패턴까지만 찾음
print(va.match('<<>aaaaa>>'))
print(va.match('0a')) # 앞자ㅔva.match('0a') # match는 첫번째 글자부터 매치 여부를 판단,
print(va.findall('akb')) # 두번쨰 글자가 패턴에 없으므로 첫번째 문자 a만 매치로 판단
```

```python
#search : 전체 글자에서 첫번째 패턴까지만 찾음
print(va.search('1aaaaa') )  # search는 전체적으로 매치되는지를 찾음
print(va.search('1aa1aaa') ) # 하지만 첫번째 패턴만 매치로 판단
```

```python
# findall

print(va.findall('a1a1ak1a')) # 전체 문자열에서 패턴에 일치하는 부분만 각각 출력

# finditer
print(va.finditer('a1a1ak1a')) # findfiter는 iterable한 객체로 저장
result = va.finditer('a1a1ak1a') 
for i in result: # result의 각 원소를 반환
    print(i)
```

```python
# 여러 문자열
text = '''student-1-:jame.student
student-2-
student-3-
student-4-
lecturer-1-
'''
p = re.compile('^student')  # student로 시작하는 첫 줄의 첫 student만 출력
result = p.findall(text) 
print(result)

p = re.compile('^student.*') # student로 시작하고 그 후에 0개 이상의 모든 문자(첫 줄만)
result = p.findall(text)     # .*: 모든 문자 1개 + 반복 0번 이상 = 모든 글자
print(result)

p = re.compile('^student.*', re.MULTILINE) # student로 시작하고 그 후에 0개 이상의 모든 문자( 모든 줄, \n 인식) 
result = p.findall(text)
print(result)
```

```python
from bs4 import BeautifulSoup
import re

with open('C:/Users/Affinity/Desktop/수업/modeule04 (1)/ch01/CSS_Sel.html',
          'r',encoding = 'utf-8') as f:
    soup = BeautifulSoup(f,'html.parser')
    tmp = soup.body # 바디 영역만 
    
    # 텍스트에서 태그를 지우고 문자열만 출력할 계획
    # 1. <~> 를 제거해야 함
    # 2. 꺽새 사이에 문자열 길이가 모두 다른 상황에서 꺽새를 지워야 함 = 임의 글자 . 활용
   
    tmp = str(soup.body) # 모두 문자열로 변경
    body=re.sub('<.+?>'," ",tmp) # tmp에서 꺽새와 그 안의 문자들을 빈칸으로 변경
    print(body)
    
    # 다른 방법
    for x in soup.body.strings: # html의 각 줄을 문자열로 읽음, \n은 태그에서 텍스트가 없는 부분
        print(repr(x)) # repr은 ""로 출력, 그냥 str 써도 됨. 다만 줄바꿈 표시가 안 됨
   
    for x in soup.body.strings: # html의 각 줄을 문자열로 읽음, \n은 태그에서 텍스트가 없는 부분
        print(str(x))           #     그냥 str 써도 됨. 다만 줄바꿈, '' 표시가 안 됨
        
    
    for x in soup.body.stripped_strings: # stripped_strings 는 \n을 표시하지 않음
        print(repr(x)) # \n은 태그에서 텍스트가 없는 부분
```

```python
from bs4 import BeautifulSoup
from urllib.request import *
from urllib.parse import *
from os import makedirs #  중간에 생략된 파일경로를 생성하며 폴더 생성
import os.path, time, re

proc_files = {} # 전역변수. 어떠한 함수에서도 쓸 수 있다. 단 지역변수랑 이름이 중복되면 안됨
                # 처리된 페이지를 여기에 기록할 것이다. 나중에 참조해서 처리된 페이지를 다시 작업 안하도록 할 것임
# html 내부에 있는 링크를 추출하는 함수
# a link 태그의 절대주소를 모은다
def enum_links(html, base): # base는 루트 주소
    soup = BeautifulSoup(html, 'html.parser')
    links = soup.select("link[re|='stylesheet']") # 외부 CSS 태그 걸린 파일을 불러올 때 이렇게 stylesheet 설정
    links += soup.select('a[href]') # 일반적인 tag 링크
    result = [] # 절대주소 목록
    for a in links:
        href = a.attrs['href'] # <a>의 속성 중에 href를 가리키는 객체
        url = urljoin(base,href) # 루트 주소에 상대주소를 붙여서 절대주소로 바꾸는 객체
        result.append(url)
    return result
#파일 다운 받고 저장하는 함수
def download_file(url): # url에서 파일을 다운받는다.
    o = urlparse(url) # url을 단위별로 나눔
    dir_path = 'C:/Users/Affinity/Desktop/수업' #파일 저장 위치
    savepath = dir_path + o.netloc + o.path # o는 url주소, netloc는 루트 주소, path는 루트 주소 하위의 폴더
                                            # ex) www.exam.com/a/ex 에서 앞에가 netloc, /a/ex는 path
    if re.search(r"/$", savepath): # savepath의 끝이 /로 끝나면 (즉, 뒤에 더 이상 폴더가 없으면)
        savepath += 'index.html' # 이 파일 주소를 붙여넣는다.
    savedir = os.path.dirname(savepath) # 이름
    
    if os.path.exists(savepath): return savepath # 주소에 파일 이름까지 있으면 주소 반환
    if not os.path.exists(savedir): # 디렉토리가 없으면 , 참고고 savedir는 디렉토리 이름까지만 있음
        print('mkdir=',savedir) 
        makedirs(savedir)       # 새로운 디렉토리 생성
    # 디렉토리 있으면 바로 try 실행
    try:
        print('download=',url)
        urlretrieve(url, savepath) # 해당 파일 주소에 해당파일 이름으로 받기
        time.sleep(1) # 잠깐 멈췃다 다시 시작하게 하는 코드. (1초 휴식) 오류를 방지하기 위한 것
    except:    #오류 났을 때 예외 처리 방법
        print('다운 실패:', url)
        return None

def analyze_html(url,root_url):
    savepath = download_file(url)
    if savepath is None: return
    if savepath in proc_files: return
    proc_files[savepath] = True
    print('analyze_html=',url)
    # 링크 추출
    html =open(savepath,'r',encoding='utf-8').read()
    links = enum_links(html, url)
    for link_url in links: # links는 result임. 즉, 링크들의 절대주소 목록
        if link_url.find(root_url)!=0:
            if not re.search(r'.css$',link_url): continue
        if re.search(r'.(html|htm)$',link_url):
            analyze_html(link_url, root_url)
            continue
        download_file(link_url)
    
```
