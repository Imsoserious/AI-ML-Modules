# 데이터 합치기

```python
df1=DataFrame({"key":["b","b","a","c","a","a","b"],"v1":range(7)})
df2=DataFrame({"key":["a","b","d"],"v2":range(2,5)})
print("{0}\n\n{1}\n\n{2}\n\n".format(df1,df2,pd.merge(df2,df1))) # 왼쪽 df에 오른쪽 df 추가
```

```python
pd.merge(df2,df1,how="outer") # outer이므로 c,d가 병합됨
```

```python
df1=DataFrame({"key1":["b","b","a","c","a","a","b"],"v1":range(7)})
df2=DataFrame({"key2":["a","b","d"],"v2":range(2,5)})
print("{0}\n\n{1}\n\n{2}\n\n".format(df1,df2,pd.merge(df1,df2,left_on="key1",right_on="key2")))
```

```python
print("{0}\n\n".format(pd.merge(df1,df2,left_on="key1",right_on="key2",how="outer")
     ))
```

```python
print("{0}\n\n".format(pd.merge(df1,df2,left_on="key1",right_on="key2")))
```

```python
df3=DataFrame({"key1":["a","d","c","a","b","a"],"key2":["aa","dd","cc","aa","bb","cc"],"v1":[1,2,3,4,5,6]})
df4=DataFrame({"key1":["a","d","c","a","b","d"],"key2":["aa","dd","cc","aa","bb","aa"],"v2":[11,22,33,44,55,66]})
print("{0}\n\n".format(pd.merge(df3,df4,on=["key1","key2"],how="outer"))) # 기준변수 두 개일 경우
```

```python
df1=DataFrame({"key1":["b","b","a","c","a","a","b"],"v1":range(7)})
df2=DataFrame({"v1":range(10,13),"v2":range(2,5)}, index=["a","d","b"])
pd.merge(df1,df2,left_on=["key1"],right_index=True, how="outer", sort=True) # 오른쪽 df에 키가 없으므로 인덱스를 키로 지정
```

```python
df1=DataFrame({"v1":range(7)},index=["b","b","a","c","a","a","b"])
df2=DataFrame({"v1":range(10,13),"v2":range(2,5)}, index=["a","d","b"])
df1
```

```python
df1.join(df2,lsuffix='_left',rsuffix='_right', how="outer", sort=True) # index를 키로 한 결합은 '왼쪽 df.join(오른쪽 df,suffix)'
```

```python
s1=Series([1,2,3],index=["a","c","b"])
s2=Series([1,4,5],index=["b","c","a"])
s3=Series([7,2,6],index=["d","a","f"])
pd.concat([s1,s2,s3],sort=True) # concat은 기본이 Outer 조인
s1,s2,s3
```

```python
pd.concat([s1,s2,s3],axis=1,sort=True) # 열결합하려면, 각 Series 내의 인덱스에 중복 값이 있으면 안된다.
```

```python
pd.concat([s1,s2,s3],axis=1,sort=True, join="inner")
```

```python
pd.concat([s1,s2,s3],axis=1,sort=True, join_axes=[["a","b"]])
```

```python
print("{0}\n\n{1}".format(pd.concat([s1,s2,s3],axis=0,sort=True,keys=[1,2,3]),pd.concat([s1,s2,s3],axis=1,sort=True,keys=[1,2,3])))
```

```python
df6=DataFrame({"name":["kim","lee","park","cho","kang"],"sex":[1,2,2,1,1],
               "home":["마포","노원","강동","분당","처인"],"income":[250,195,200,324,274]})
num_to_str={1:"남",2:"여"}
to_city={"마포":"서울","노원":"서울","강동":"서울","분당":"경기","처인":"수원"}
df6
```

```python
def f(x):
    return x*0.015

df6["성별"]=df6["sex"].map(num_to_str)
df6["도시"]=df6["home"].map(to_city)
df6["remium"]=df6["income"].map(f) # map에 함수 이름 넣기
df6
```

```python
age= [3,10,15,25,19,40,34,60,80,95,72,66,3,90,70]
bins= [3,20,30,50,70,90]
labels=["청소년","청년","중년","장년", "노년"]
cats=pd.cut(age,bins,labels=labels,include_lowest=True,right=False) # labels 개수는ㄴ bins-1개
cats
```

```python
dfa=DataFrame(age,columns=["age"])
dfa["age_cat"]=pd.cut(dfa.age,bins=bins,labels=labels,include_lowest=True,right=False)
dfa
```

```python
dfa=DataFrame(age,columns=["age"])
dfa["age_qcut"]=pd.qcut(dfa.age,5,labels=["청소년","청년","중년","장년", "노년"])
dfa
```
